# Comapre the Change in population of countries from 2022 and 1970

```python
df
```

```python
population_2022_1970_df=df[['Country/Territory','Capital','2022 Population','1970 Population']]
```

```python
population_2022_1970_df
```

```python
population_2022_1970_df['difference_in_population']=df['2022 Population']-df['1970 Population']
```

```python
population_2022_1970_df
```

```python
population_2022_1970_df['Popuation_growth_rate']=(population_2022_1970_df['difference_in_population']/population_2022_1970_df['1970 Population'])*100
```

```python
population_2022_1970_df
```

```python
population_2022_1970_df=population_2022_1970_df.sort_values(by="Popuation_growth_rate")
```

```python
population_2022_1970_df
```

```python
population_2022_1970_df=population_2022_1970_df.sort_values(by="Popuation_growth_rate",ascending=False)
```

```python
population_2022_1970_df
```

```python
plt.figure(figsize=(20,20))
sns.barplot(x=population_2022_1970_df['Country/Territory'],y=population_2022_1970_df['Popuation_growth_rate'])
plt.xticks(rotation=90)
plt.show()
```
